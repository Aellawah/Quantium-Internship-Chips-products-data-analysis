## Quantium Internship Module 1:Chips products data analysis

**About the dataset**

Our client is a company that owns several hypermarkets in the region that sells different types of products, the client wants to better understand the types of customers who purchase **Chips** and their purchasing behaviour within the region.

The client has provided us with two datasets related to chips transactions:

- Transactions dataset
- Customers dataset

**Transactions** dataset Features description:

- **DATE:** The date of the transaction
- **STORE_NBR:** The store number in the system from various stores numbers
- **LYLTY_CARD_NBR:** The customer's card that was used during the transaction that shows his purchasing history
- **TXN_ID:** Transaction ID
- **PROD_NBR:** The number of the product in the system
- **PROD_NAME:** The name of the product
- **PROD_QTY:** The quantity that the customer purchased during the transaction
- **TOT_SALES:** Total sales amount of the transaction

**Customers** dataset Features description:

- **LYLTY_CARD_NBR:**The customer's card that was used during the transaction that shows his purchasing history
- **LIFESTAGE:** Customer attribute that identifies whether a customer has a family or not and what point in life they are at e.g. are their children in pre-school/primary/secondary school
- **PREMIUM_CUSTOMER:**Customer segmentation used to differentiate shoppers by the price point of products they buy and the types of products they buy. It is used to identify whether customers may spend more for quality or brand or whether they will purchase the cheapest options

```python
#converting dates to datetime in pandas
def convert_to_datetime(num):
    dt = datetime.fromordinal(datetime(1900, 1, 1).toordinal() + num - 2)
    return dt
```

### Features Engineering

Now let's extract important features from the data to consider in our analysis such as:

1. Packet size column
2. Year column

```python
Transactions.head(2)
```

Splitting **packet size** from product name

```python
#creating a function to split packet size from product name
def packet_size(Transactions):
    string = Transactions["PROD_NAME"]
    num = []
    for i in string:
        if i.isdigit():
            num.append(i)
    number = "".join(num)
    return int(number)
```

```python
#applying the function on product name column
Transactions["PACKET_SIZE"] = Transactions.apply(packet_size,axis=1)
Transactions.head()
```

Extracting from **DATE** column **year,month,week_day_name,month_day_name** columns

```python
# Extracting features
Transactions['Transaction_year']=Transactions['DATE'].dt.year
Transactions['Transaction_month']=Transactions['DATE'].dt.month_name()
Transactions['Transaction_month_day']=Transactions['DATE'].dt.day
Transactions['Transaction_week_day']=Transactions['DATE'].dt.day_name()
```

```python
# Ammending datatypes
Transactions['Transaction_year']=Transactions['Transaction_year'].astype(str)
Transactions['Transaction_month_day']=Transactions['Transaction_month_day'].astype(str)
```

Done with the extractions now let's Check for outliers

```python
# A function that will give us the boundries of the data which we should remove outliers after
def outlier_treatment(datacolumn):
    sorted(datacolumn)
    Q1,Q3 = np.percentile(datacolumn , [25,75])
    IQR = Q3 - Q1
    lower_range = Q1 - (1.5 * IQR)
    upper_range = Q3 + (1.5 * IQR)
    return lower_range,upper_range
```

```python
#Checking for outliers in product quantity column
plt.figure(figsize=(8,4),dpi=100);
sns.boxplot(Transactions['PROD_QTY'],color=base_color);
```

```python
outlier_treatment(Transactions['PROD_QTY'])
```

we will remove the data after value (2) for this column

```python
#Checking for outliers in total sales quantity column
plt.figure(figsize=(8,4),dpi=100);
sns.boxplot(Transactions['TOT_SALES'],color=base_color);
```

```python
outlier_treatment(Transactions['TOT_SALES'])
```

we will remove data after 15 for this column

```python
#Checking for outliers in product size column
plt.figure(figsize=(8,4),dpi=100);
sns.boxplot(Transactions['PACKET_SIZE'],color=base_color);
```

```python
outlier_treatment(Transactions['PACKET_SIZE'])
```

It seems that we have outliers more than 212 and less than 111

```python
#Removing outliers
Transactions = Transactions[Transactions["PROD_QTY"] <=2]
Transactions = Transactions[Transactions["TOT_SALES"] <=15]
Transactions = Transactions[(Transactions["PACKET_SIZE"] <=213) & (Transactions["PACKET_SIZE"] >=111)]
```
